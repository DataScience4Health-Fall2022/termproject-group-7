### Step 3: Missing Data Analysis 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

a = pd.read_csv('covidfulldata.csv', index_col=[0,1,2,3,4,5,6])

plt.figure(figsize = (20,20))
plt.xlabel("participants (n=49968)")
plt.ylabel("Features")
s = sns.heatmap(a.isna().transpose(),
           cmap="rocket_r",
           cbar_kws={'label': 'Missing data'},
           xticklabels=False)
s.set(xlabel='Participants (n=49968)', ylabel='Features')
```

#### Analysis
We can see that from the plot above, missing data is distributed mainly in questions for sex other input and employment status other input. We have also found a large proportion of missing data for Cognitive Reflection Test (CRT) section. The rationale behind this might be that this section requires text inputs from users and this section resides in the middel to ending part of the whole questionnaire thus most users chose to skip these questions.

#### Decision
- We would like to include the country column into the data set and remove the tuples with None for country.
- We would like to remove the CRT section (3 questions - 3 columns) as there is no meaning of computing the mean values given the large number of missing values.

### Step 5: Imputation

We have tried the 2 simple techniques for imputation:
- Using Mean value of the corresponding feature column
- Using Median value of the corresponding feature column 

#### Imputation with Mean and Median in one Method

```python
import numpy as np
import statistics as st

def process(data, data_mean, data_median, method):
    for i in range(1, len(data)):
        for j in range(7, len(data[0])):
            if data[i][j] == 'NA' or data[i][j] == '':
#                 print("here")
                if method == 'mean':
                    data[i][j] = data_mean[j]
                elif method == 'median':
                    data[i][j] = data_median[j]
                    
def get_mean_median(data):
    data_mean = [0 for i in range(len(data[0]))]
    data_median = [0 for i in range(len(data[0]))]
        
    for j in range(7, len(data[0])):
        num = []
        for i in range(1, len(data)):
            if isinstance(data[i][j], str) and data[i][j].isnumeric():
                num.append(float(data[i][j]))
#         print(data[0][j])
        mean = st.mean(num)
        median = st.median(num)
#         print(mean)
#         print(median)
        data_mean[j] = mean
        data_median[j] = median
    return data_mean,data_median
                 
```

```python
imputation_data_mean = data_categorical.copy()
imputation_data_median = data_categorical.copy()
res= get_mean_median(imputation_data_mean)
print(res)
print(len(res[0]))
print(len(res[1]))

# process(imputation_data_mean, data_mean, data_median, 'mean')
# process(imputation_data_median, data_mean, data_median, 'median')
```

### Step 7: Data Balancing

```python
num = 0
for d in data_final:
    if d[-2] == 2:
        num += 1
#         print("here")
print("There are ", num, " positive cases within ", len(data) - 1, " pieces of data.")
```

```python
data_final_final = []
data_final_final.append(data_final[0])
num1 = num
for d in data_final[1:]:
    if(d[-2] == 1 and num1 > 0):
        d[-2] = 0
        data_final_final.append(d)
        num1 -= 1
    elif(d[-2] == 2):
        d[-2] = 1
        data_final_final.append(d)

print(len(data_final_final))
        
    
```

## Models

### Model 1: SVM

```python
data_data = [d[:-2] for d in data_final_final][1:]
target_positive = [d[-2] for d in data_final_final][1:]
target_known_positive = [d[-1] for d in data_final_final][1:]
```

```python
from sklearn.model_selection import train_test_split

# 80% training 20% testing 
X_train, X_test, y_train, y_test = train_test_split(data_data, target_positive, test_size=0.2,random_state=109) 
```

```python
from sklearn import svm
import time

start_time = time.time()
clf = svm.SVC(kernel='linear')
clf.fit(X_train, y_train)
print("Training for SVM used ", time.time() - start_time, " seconds.")

y_pred = clf.predict(X_test)
```

### Evaluation

```python
from sklearn import metrics
from sklearn.metrics import f1_score

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
f1_score(y_test, y_pred)
```

```python
def evaluate(y_actual,y_pred):
    ## Your code here
    false_positive = 0
    false_negative = 0
    true_positive = 0
    true_negative = 0
    for i in range(0, len(y_actual)):
        if y_actual[i] == y_pred[i]:
            if y_pred[i] == 1:
                true_positive += 1
            else:
                true_negative += 1
        else:
            if y_pred[i] == 1:
                false_positive += 1
            else:
                false_negative += 1
    
    accuracy = (true_positive + true_negative)/len(y_actual)
                
    return false_positive, false_negative, true_positive, true_negative, accuracy
```

```python
res = evaluate(y_test, y_pred)
print("Accuracy: ", res[4])
print("matrix")
print("                 pred_positive      pred_negative")
print("actual_positive      ", res[2], "               ", res[1])
print("actual_negative      ", res[0], "               ", res[3])
```
